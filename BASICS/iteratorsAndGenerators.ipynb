```python
class iterator:
    def __init__(self,val):
        self.val = val
    def __iter__(self):
        self.num = 1
        return self
    def __next__(self):
        if self.num <= self.val:
            final = self.num*5
            self.num += 1
            return final
        else:
            raise StopIteration
obj = iterator(5)
itr = iter(obj)
print(next(itr))
```
